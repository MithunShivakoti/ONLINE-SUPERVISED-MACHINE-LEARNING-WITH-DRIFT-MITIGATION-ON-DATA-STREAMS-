```python
class_order = [0,1,2]
class_mapping = {class_label: i for i, class_label in enumerate(class_order)}

df_sorted = data_ac.copy()
df_sorted['class_id'] = df_sorted['class'].map(class_mapping)
df_sorted.sort_values(by='class_id', inplace=True)
df_sorted.drop(columns=['class_id'], inplace=True)
df_sorted.reset_index(drop=True, inplace=True)
```

```python
data=pd.read_csv("/content/drive/MyDrive/traffic/tfidf_embeddings.csv")
data
```

```python
df_sorted = df_sorted.apply(pd.to_numeric, errors='coerce')
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
X_train = data.iloc[:, :-1]
y_train = data.iloc[:, -1]
X_test = df_sorted.iloc[:,:-1]
y_test = df_sorted.iloc[:,-1]
```

```python
from river import drift
drift=[drift.KSWIN,drift.PageHinkley,drift.ADWIN]
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
svm_classifier = SVC(kernel='rbf', C=1.0)  # You can choose different kernels (e.g., 'linear', 'rbf', 'poly')
def mod(X_train,y_train):

  # Fit the SVM model to the training data
  svm_classifier.fit(X_train, y_train)
```

```python
df_modified = df_sorted.copy()
df_modified.loc[df_modified['class'].isin([0, 1, 2]), 'class'] = 0
```

```python
for i in range(3):
  begin = time.time()
  detector = drift[i]()
  index = 0
  k = 1
  c = 1
  X_train = data.iloc[:, :-1]
  y_train = data.iloc[:, -1]
  mod(X_train,y_train)
  print(drift[i])
  for j in range(y_test.shape[0]):
    row = X_test.iloc[j]
    row_np = np.array(row)
    row_statistic = np.var(row)
    detector.update(row_statistic)
    if detector.change_detected:
      print(f'Drift detected at index {k}')
      df_modified.loc[index:, 'class'] = c
      c+=1
      X_train = df_modified.iloc[:, :-1]
      y_train = df_modified.iloc[:, -1]

      mod(X_train,y_train)
      detector.reset()
    index+=1
    k+=1
  end = time.time()
  print("Time taken for",drift[i],end-begin)
  X = df_modified.iloc[:,:-1]
  y = df_modified.iloc[:,-1]
  X_train, X_test, y_train, y_test = tts(X, y, test_size=0.33, random_state=42,shuffle = True)
  svm_classifier.fit(X_train, y_train)

  # Predict the labels for the test data
  y_predict = svm_classifier.predict(X_test)

  # Calculate accuracy
  svm_score = accuracy_score(y_test, y_predict)
  print('Accuracy of SVM: ', svm_score)

  # Calculate precision, recall, and F1-score
  precision, recall, fscore, support = precision_recall_fscore_support(y_test, y_predict, average='weighted')
  print('Precision of SVM: ', precision)
  print('Recall of SVM: ', recall)
  print('F1-score of SVM: ', fscore)
  print("Training scores")
  y_predict = svm_classifier.predict(X_train)

  # Calculate accuracy
  svm_score = accuracy_score(y_train, y_predict)
  print('Accuracy of SVM: ', svm_score)

  # Calculate precision, recall, and F1-score
  precision, recall, fscore, support = precision_recall_fscore_support(y_train, y_predict, average='weighted')
  print('Precision of SVM: ', precision)
  print('Recall of SVM: ', recall)
  print('F1-score of SVM: ', fscore)
```
