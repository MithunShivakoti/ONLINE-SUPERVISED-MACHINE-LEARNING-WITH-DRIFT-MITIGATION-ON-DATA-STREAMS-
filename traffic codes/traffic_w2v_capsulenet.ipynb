```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
```

```python
data = pd.read_csv("/content/drive/MyDrive/traffic/word2vec_embeddings_3.csv")
data
```

```python
X=data.drop(["class"], axis=1)
y=data["class"]
```

```python
vec, vec_test, label, label_test = tts(X, y, test_size=0.20, random_state=42,shuffle = True)
print(vec.shape)
print(label.shape)
print(vec_test.shape)
print(label_test.shape)
vec_test = np.array(vec_test)
vec = np.array(vec)
label=label.astype(np.float32)
label_test=label_test.astype(np.float32)
```

```python
vec_test = np.array(vec_test).astype(np.float32)
vec = np.array(vec.reshape(-1,4,25,1)).astype(np.float32)
vec_test = vec_test.reshape(-1,4,25,1).astype(np.float32)
print(vec.shape)
print(label.shape)
print(vec_test.shape)
print(label_test.shape)
```

```python
caps_model = CapsNet()
earlyStopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2,patience=3, min_lr=1e-5)
#tensorBoard = tf.keras.callbacks.TensorBoard(log_dir='logs')
caps_model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),loss=caps_model.margin_loss,metrics=['accuracy'])
caps_model.fit(vec,label,batch_size=32,epochs=50,validation_data=(vec_test,label_test),callbacks=[earlyStopping,reduce_lr])
```
