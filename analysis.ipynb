# WE ARE NOW WORKING ON DATASET CLASS 2

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud


# Load the CSV file into a DataFrame
df = pd.read_csv("C://Users//shree//Desktop//VIT AP//Capstone//PROJECT//DATA SET//1_TrainingSet_2Class.csv")
print(df.head()) 
print(df.info())  
```

```python
df = df.iloc[:, 2:] 
df.head()
```

```python
column_names = ['tweets']
df.columns=column_names
df.head
```

```python
df.shape
```

```python
tweets = df['tweets']

# Word Cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(tweets))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Tweets')
plt.show()

# Bar plot of top N frequent words
from collections import Counter
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
words = ' '.join(tweets).lower()
word_tokens = word_tokenize(words)
filtered_words = [word for word in word_tokens if word.isalnum() and word not in stop_words]
word_freq = Counter(filtered_words)

top_n = 50
top_words = [word[0] for word in word_freq.most_common(top_n)]
top_word_counts = [word[1] for word in word_freq.most_common(top_n)]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_word_counts, y=top_words)
plt.title(f'Top {top_n} Most Frequent Words in Tweets')
plt.xlabel('Frequency')
plt.ylabel('Word')
plt.show()

# Additional plots (you can customize these based on your needs)
# For example:
# - Histogram of tweet lengths
# - Sentiment analysis distribution
# - Time series analysis of tweet volume

# Don't forget to install the necessary packages if you haven't already:
# pip install pandas matplotlib seaborn nltk wordcloud
```

```python
df['tweet_length'] = df['tweets'].apply(len)
#here we will calculate the tweet length and then add another column which shows the length of each tweet.

print(df.head())
```

```python
import matplotlib.pyplot as plt

# Create a histogram of tweet lengths
plt.figure(figsize=(10, 6))
plt.hist(df['tweet_length'], bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Tweet Lengths')
plt.xlabel('Tweet Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
from textblob import TextBlob
# Function to get sentiment polarity using TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Apply sentiment analysis to each tweet and create a new column
df['sentiment'] = df['tweets'].apply(get_sentiment)

# Display the updated DataFrame
print(df.head())
```

```python

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment'], bins=20, color='green', kde=True)
plt.title('Sentiment Analysis Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
